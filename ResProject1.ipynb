```python
import pandas as pd

input_file = "D:/DATA FOR RESEARCH PROJECT/deviceData_yearly.csv"
output_file = "D:/DATA FOR RESEARCH PROJECT/pacemaker_filtered.csv"

chunksize = 1000000  # You can reduce this if memory usage is still too high
pacemaker_chunks = []

for i, chunk in enumerate(pd.read_csv(input_file, chunksize=chunksize, encoding='latin1', dtype=str)):
    # Convert to lowercase and check for 'pacemaker' in GENERIC_NAME
    filtered_chunk = chunk[chunk['GENERIC_NAME'].str.lower().str.contains('pacemaker', na=False)]
    
    if not filtered_chunk.empty:
        pacemaker_chunks.append(filtered_chunk)
        print(f"âœ… Chunk {i+1}: Found {filtered_chunk.shape[0]} pacemaker-related rows.")

# Combine and save
if pacemaker_chunks:
    result_df = pd.concat(pacemaker_chunks, ignore_index=True)
    result_df.to_csv(output_file, index=False)
    print(f"Saved filtered pacemaker data: {output_file} ({result_df.shape[0]} rows)")
else:
    print(" No rows with 'pacemaker' found in GENERIC_NAME.")
```

```python
import pandas as pd
import csv

# Load pacemaker_filtered CSV (small enough to load once)
pacemaker_path = 'D:/DATA FOR RESEARCH PROJECT/pacemaker_filtered.csv'
pacemaker_df = pd.read_csv(pacemaker_path, dtype=str)
print(f"Pacemaker records: {pacemaker_df.shape}")

# Set of keys to speed up lookup
pacemaker_keys = set(pacemaker_df['MDR_REPORT_KEY'])

# Define file paths
master_path = 'D:/DATA FOR RESEARCH PROJECT/Extracted files/mdrfoiThru2024.txt'
output_path = 'D:/DATA FOR RESEARCH PROJECT/master_pacemaker_joined.csv'

# Open output file and write headers
write_header = True
chunksize = 100000

# Process master file in chunks
chunk_iter = pd.read_csv(
    master_path,
    delimiter='|',
    encoding='latin1',
    dtype=str,
    engine='python',
    chunksize=chunksize,
    quoting=csv.QUOTE_NONE,
    on_bad_lines='skip'
)

for i, chunk in enumerate(chunk_iter):
    print(f"ðŸ”„ Processing chunk {i+1}...")

    # Filter rows in chunk that exist in pacemaker_filtered
    chunk_filtered = chunk[chunk['MDR_REPORT_KEY'].isin(pacemaker_keys)]

    # Merge the filtered chunk with pacemaker_df
    merged_chunk = chunk_filtered.merge(pacemaker_df, on='MDR_REPORT_KEY', how='inner')

    # Save to CSV (append mode)
    merged_chunk.to_csv(output_path, index=False, mode='a', header=write_header)
    write_header = False  # Only write header for first chunk

    print(f"âœ… Chunk {i+1} processed. Rows merged: {merged_chunk.shape[0]}")

print("All chunks processed and merged successfully!")
```
